```python
data = pd.read_csv('2019_nCoV_data.csv', delimiter = ',')
data.head()
```

```python
data.info()
```

```python
adjusted_unique_dates = list(data['Last Update'].unique())
adjusted_unique_dates.sort()
# adjusted_unique_dates
```

```python
fix_date = []
for i in adjusted_unique_dates[-175:]:
    d = datetime.datetime.strptime(i, '%Y-%d-%m %H:%M:%S')
    convert = datetime.datetime.strftime(d, "%Y-%m-%d %H:%M:%S")
    fix_date.append(convert)
    data.replace(i, convert, inplace=True)
adjusted_unique_dates[-175:] = fix_date
```

```python
adjusted_unique_dates.sort()
# adjusted_unique_dates
```

```python
fix_date_2 = []
for i in adjusted_unique_dates[20:33]:
    d2 = datetime.datetime.strptime(i, '%Y-%d-%m %H:%M:%S')
    convert2 = datetime.datetime.strftime(d2, "%Y-%m-%d %H:%M:%S")
    fix_date_2.append(convert2)
    data.replace(i, convert2, inplace=True)
    
adjusted_unique_dates[20:35] = fix_date_2
```

```python
adjusted_unique_dates.sort()
# adjusted_unique_dates
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.replace('China', 'Mainland China', inplace=True)
```

```python
excess_dates = []
filtered_unique_dates = copy.deepcopy(adjusted_unique_dates)
for i in range(len(filtered_unique_dates)):
    # assume the number of coronavirus cases, deaths, and recover increases over time 
    if i != 0:
        current_day_cases = data[data['Last Update']==filtered_unique_dates[i]].Confirmed.sum()
        prev_day_cases = data[data['Last Update']==filtered_unique_dates[i-1]].Confirmed.sum()
        current_day_deaths = data[data['Last Update']==filtered_unique_dates[i]].Deaths.sum()
        prev_day_deaths = data[data['Last Update']==filtered_unique_dates[i-1]].Deaths.sum()
        current_day_recovered = data[data['Last Update']==filtered_unique_dates[i]].Recovered.sum()
        prev_day_recovered = data[data['Last Update']==filtered_unique_dates[i-1]].Recovered.sum()
        
        if(current_day_cases < prev_day_cases or current_day_deaths < prev_day_deaths or current_day_recovered < prev_day_recovered):
            excess_dates.append(adjusted_unique_dates[i])
            # swap the current date with the previous date, it will get removed later
            temp = filtered_unique_dates[i]
            filtered_unique_dates[i] = filtered_unique_dates[i-1]
            filtered_unique_dates[i-1] = temp
            
for i in excess_dates:
    filtered_unique_dates.remove(i)
```

```python
world_cases = []
deaths = [] 
mortality_rate = []
recovered = [] 

for i in filtered_unique_dates:
    confirmed_sum = data[data['Last Update']==i].Confirmed.sum()
    death_sum = data[data['Last Update']==i].Deaths.sum()
    recovered_sum = data[data['Last Update']==i].Recovered.sum()
    world_cases.append(confirmed_sum)
    deaths.append(death_sum)
    mortality_rate.append(death_sum/confirmed_sum)
    recovered.append(recovered_sum)
```

```python
plt.figure(figsize=(20, 12))
plt.plot(filtered_unique_dates, world_cases)
plt.title('# of Coronavirus Cases Over Time', size=30)
plt.xlabel('Time', size=30)
plt.ylabel('# of Cases', size=30)
plt.xticks(rotation=50, size=15)
plt.show()
```

```python
plt.figure(figsize=(20, 12))
plt.plot(filtered_unique_dates, deaths, color='red')
plt.title('# of Coronavirus Deaths Over Time', size=30)
plt.xlabel('Time', size=30)
plt.ylabel('# of Deaths', size=30)
plt.xticks(rotation=50, size=15)
plt.show()
```

```python
mean_mortality_rate = np.mean(mortality_rate)
plt.figure(figsize=(20, 12))
plt.plot(filtered_unique_dates, mortality_rate, color='orange')
plt.axhline(y = mean_mortality_rate,linestyle='--', color='black')
plt.title('Mortality Rate of Coronavirus Over Time', size=30)
plt.legend(['mortality rate', 'y='+str(mean_mortality_rate)])
plt.xlabel('Time', size=30)
plt.ylabel('Mortality Rate', size=30)
plt.xticks(rotation=50, size=15)
plt.show()
```

```python
unique_provinces =  data['Province/State'][data.Confirmed > 0].unique()
unique_provinces
```

```python
province_confirmed_cases = []
for i in unique_provinces:
    province_confirmed_cases.append(data[data.Confirmed>0][data['Province/State']==i].Confirmed.max())
```

```python
nan_indices = [] 

# handle nan if there is any, it is usually a float: float('nan')
for i in range(len(unique_provinces)):
    if type(unique_provinces[i]) == float:
        nan_indices.append(i)

unique_provinces = list(unique_provinces)
province_confirmed_cases = list(province_confirmed_cases)

for i in nan_indices:
    unique_provinces.pop(i)
    province_confirmed_cases.pop(i)
```

```python
# number of cases per country/region
unique_countries = data[data.Confirmed>0]['Country'].unique()
unique_countries.sort()
unique_countries
```

```python
country_confirmed_cases = []
latest_date = filtered_unique_dates[-1]
for i in unique_countries:   
    if i == 'Mainland China' or i == 'US' or i == 'Canada' or i == 'Australia':
        c = 0
        provinces = data[data.Country==i]['Province/State'].unique()
        filtered_provinces = []
        
        # remove null values from provinces
        for p in provinces:
            if type(p) != float:
                filtered_provinces.append(p)
                
        for province in filtered_provinces:
            index = -1
            while(True):
                if(len(data[data['Province/State']==province][data['Last Update']==adjusted_unique_dates[index]])>0):
                    #print(data[data['Province/State']==province][data['Last Update']==adjusted_unique_dates[index]].Confirmed.max())
                    c +=data[data['Province/State']==province][data['Last Update']==adjusted_unique_dates[index]].Confirmed.max()
                    break
                else:
                    index -= 1  
        country_confirmed_cases.append(c) 

    else:
        index = -1
        while(True):
            if(len(data[data['Country']==i][data['Last Update']==adjusted_unique_dates[index]])>0):
                country_confirmed_cases.append(data[data['Country']==i][data['Last Update']==adjusted_unique_dates[index]].Confirmed.max()) 
                break
            else:
                index -= 1
```
