```python
import numpy as np
```

```python
t = np.arange(0,2,0.01)
s = np.sin(2 * np.pi * t + np.pi/6)
```

```python
fig, ax = plt.subplots()
ax.plot(t, s, color='red')
ax.grid()


ax.set(xlabel='X', ylabel='Y',
       title='First Plot', )

plt.show()
```

```python
t = np.random.normal(10,150,100)
s = (3*t**2+1)
```

```python
fig, ax = plt.subplots()
ax.scatter(t, s, color='red')
ax.grid()


ax.set(xlabel='X', ylabel='Y',
       title='Seccond Plot', )

plt.show()
```
